## indexing, slicing

* `iloc`: 정수 인덱스(ndarray의 기본 인덱스) 기반으로 원소 참조 또는 슬라이싱. (integer location)
    * 시작점에서 인덱스 시작. -> 슬라이싱이 달라진다.
* `loc`: 레이블 기반의 원소 참조 또는 슬라이싱. (location)
    * 레이블: 포스트잇 형식. -> 슬라이싱이 달라진다.

```python
np.random.seed(1)
s1 = pd.Series(data=np.random.randint(-10, 10, size=6))
s1
```

```python
# iloc를 사용한 인덱싱.
print(s1.iloc[0]) # Series에서 첫번째 원소.
print(s1.iloc[-1]) # Series에서 마지막 원소.
```

```python
# iloc를 사용한 슬라이싱.
print(s1.iloc[:3]) # Series에서 첫 3개 원소 자르기.
print(s1.iloc[-3:]) # Series에서 마지막 3개 원소 자르기.
```

```python
# loc를 사용한 indexing
print(s1.loc[0]) # Series의 첫번째 원소.
print(s1.loc[-1]) # Series의 마지막 원소. -> error가 난다.(ValueError: -1 is not in range) -> 음수 인덱스 사용 불가능.
# 옆에 있는 인덱스를 그대로 사용해야된다. (포스트잇을 생각하면 -1이라는 포스트잇은 존재하지 않는다.)
```

```python
# loc를 사용한 슬라이싱.
print(s1.loc[0:2])
```

```python
print(s1.loc[:2])
```

# DataFrame에서 조건의 맞는 행을 선택 - boolean indexing

```python
df
```

```python
df.year == 2023
```

```python
df[df.year == 2023]
```

* boolean indexing 을 사용할 때는, 파이썬의 논리 연산자(`and, or, not`)를 사용할 수 없다.
* boolean indexing에서는 `&, |, ~` 연산자를 사용한다.
* 조건식과 조건식은 `()`를 사용해서 구별한다.

```python
# city가 '서울'이거나, pop이 0.6 이상인 행들을 선택:
df[(df['city'] == '서울') | (df['pop'] <= 0.65)]
```

```python
# 경기도의 2023년 자료를 선택
df[(df.city == '경기') & (df.year == 2023)] # 왼쪽을 검사하고 -> 왼쪽으로 간다.
```

```python
math_mean = exam['math'].describe()[1]
print(f'수학 평균 = {math_mean}')
english_mean = exam['english'].describe()[1]
print(f'영어 평균 = {english_mean}')
science_mean = exam['science'].describe()[1]
print(f'과학 평균 = {science_mean}')
```

```python
for x in range(1, 6):
    math_mean_class = exam[exam['class'] == x].math.mean()
    print(f'class {x} : {math_mean_class}')
```

```python
exam_dict = {'class' : np.arange(1, 6)}
means = []
for x in range(1, 6):
    math_mean_class = exam[exam['class'] == x].math.mean()
    means.append(math_mean_class)

print(means)

exam_dict = {
    'class': np.arange(1, 6),
    'math_mean': means
}
print(exam_dict)

df = pd.DataFrame(exam_dict)
df
```

# 1반 학생들의 수학 평균 점수 이상인 학생들

```python
class_one_math_mean = df[df['class'] == 1]['math_mean']
math_mean_class1 = exam[exam['class'] == 1].math.mean()
print(math_mean_class1)
print(class_one_math_mean)
exam[exam['math'] >= float(class_one_math_mean)]
```
