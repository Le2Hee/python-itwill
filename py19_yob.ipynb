```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

* YOB: Year of Birth
* github = https://github.com/wesm/pydata-book/raw/3rd-edition/datasets/babynames/yob1880.txt
* 1880년 ~ 2010년까지 연도별로 텍스트(txt) 파일이 저장되어 있다.
* 각 파일에는 '이름,성별,신생아수'가 comma로 구분되어 있다.
* (주의) 컬럼 이름은 파일에 없다.

```python
file_path = 'https://github.com/wesm/pydata-book/raw/3rd-edition/datasets/babynames/yob1880.txt'
```

```python
yob1880 = pd.read_csv(file_path)
```

```python
yob1880.head()
```

`pd.read.csv()` 함수는 파일의 첫번째 줄을 데이터프레임의 컬럼 이름으로 취급함.

CSV 파일에서 컬럼 이름이 포함되어 있지 않은 경우 (파일의 첫번째 줄부터 데이터가 시작되는 경우)에는 컬럼 이름들을 함수 호출할 때 아규먼트로 전달해야 한다.

```python
col_names = ['name', 'gender', 'births', 'year'] # 데이터프레임 컬럼 이름 리스트.
yob1880 = pd.read_csv(file_path, names=col_names)
```

```python
yob1880.head()
```

1880년 ~ 2010년까지 각각의 txt 파일들을 읽어서 만들어진 연도별 데이터프레임들을 저장하는 리스트 만들기.
(결과) 131개의 데이터프레임을 저장하는 리스트가 만들어져야 한다.

각각의 데이터 프레임에는 year 컬럼을 추가.

```python
yobs = []
col_names = ['name', 'gender', 'births', 'year']
for year in range(1880, 2011):
    file_path = f'https://github.com/wesm/pydata-book/raw/3rd-edition/datasets/babynames/yob{year}.txt'
    df = pd.read_csv(file_path, names=col_names)
    df['year'] = year
    yobs.append(df)
```

```python
yobs
```

```python
len(yobs) # 2010 - 1880 + 1 = 131
```

```python
# 리스트의 첫번째 데이터 프레임
yobs[0].head()
```

```python
yobs[0].tail()
```

```python
# 리스트에서 가장 마지막 데이터프레임
yobs[-1].head()
```

131개의 데이터프레임들을 1개의 데이터프레임으로 합치기.

```python
yob_df = pd.concat(yobs, ignore_index=True)
# ignore_index: 기본값은 Fasle이고, True인 경우에는 각 데이터프레임들의 기존 인덱스를 무시하고(ignore), 재색인(re-indexing)을 한다.
# axis: axis=0 -> 세로, axis=1 -> 가로. 방향으로 프레임을 붙이겠다는 소리. (기본값 axis=0)
```

```python
yob_df.head()
```

```python
yob_df.tail()
```

# 연도별 신생아 수, 시각화

```python
count_year = yob_df.year.value_counts().index
count_year_ba = yob_df.year.value_counts().values
sns.barplot(x=count_year, y=count_year_ba)
plt.show()
```

```python
births_by_year = yob_df.groupby('year').births.sum()
plt.plot(births_by_year.index, births_by_year.values)
plt.grid()
plt.xlabel('year')
plt.ylabel('# of birth')
plt.show()
```

```python
births_by_year = yob_df.groupby('year').births.sum()
plt.bar(births_by_year.index, births_by_year.values)
plt.grid()
plt.xlabel('year')
plt.ylabel('# of birth')
plt.show()
```

```python
sns.lineplot(x=births_by_year.index, y=births_by_year.values)
plt.grid()
plt.xlabel('year')
plt.ylabel('# of birth')
plt.show()
```

```python
sns.lineplot(data=yob_df, x='year', y='births', estimator='sum')
plt.grid()
plt.show()
```

# 연도별 성별 신생아 수 찾고, 다양하게 시각화

```python
births_by_year_gender = yob_df.groupby(['year', 'gender']).births.sum()
births_by_year_gender
```

```python
sns.lineplot(data=births_by_year_gender, x='year', y=births_by_year_gender.values, hue='gender')
plt.grid()
plt.show()
```

```python
births_pivot = yob_df.pivot_table(values='births', index='year', columns='gender', aggfunc='sum')
births_pivot
```

```python
births_pivot.index
```

```python
plt.plot(births_pivot.index, births_pivot.F, label='Female')
plt.plot(births_pivot.index, births_pivot.M, label='Male')

plt.legend()
plt.grid()

plt.show()
```

```python
df = births_by_year_gender.reset_index()
# reset_index(): Series, DataFrame의 인덱스들을 DataFrame의 컬럼으로 만들어 준다.
df
```

```python
sns.lineplot(data=yob_df, x='year', y='births', hue='gender', estimator='sum', errorbar=None)
plt.grid()
plt.show()
```
