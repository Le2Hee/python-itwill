## 막대 그래프

### `countplot()` 함수

카테고리 타입 변수의 빈도수를 막대로 표현한 그래프.

```python
# 성별 빈도수, 세로 막대 그래프를 countplot을 이용해서 그리기.
sns.countplot(data=tips, x='sex')
plt.show()
```

```python
# 성별 빈도수 가로 막대 그래프.
sns.countplot(data=tips, y='sex')
plt.show()
```

```python
# 요일 빈도수 막대 그래프
sns.countplot(data=tips, x='day')
plt.show()
```

```python
# 요일별 성별 빈도수
tips.groupby(['day', 'sex']).tip.count()
```

```python
sns.countplot(data=tips, x='day', hue='sex')
# hue 파라미터: 그룹 별로 다른 색깔을 설정한다.
plt.show()
```

```python
sns.countplot(data=tips, x='sex', hue='day')
plt.show()
```

```python
sns.countplot(data=tips, x='day', hue='smoker')
plt.show()
```

```python
# 성별, 흡연여부별 빈도수
sns.countplot(data=tips, x='sex', hue='smoker')
plt.show()
```

## `barplot()` 함수

* 그룹(카테고리) 별로 변수의 통계량(평균, 표준편차, 중앙값, ...)을 막대 그래프로 표현.
* 통계량 계산 함수의 기본값은 평균(mean).
* 카테고리 타입 변수와 숫자 타입 변수를 함께 시각화.


```python
# 성별 tip의 평균
tips.groupby('sex').tip.mean()
```

```python
sns.barplot(data=tips, x='sex', y='tip') # estimator = `mean` : 기본값, 평균 계산
plt.show()
```

```python
sns.barplot(data=flights, x='month', y='year')
plt.show()
```

```python
tips.groupby('sex').tip.median() # 중위값, 중앙값, 50%
```

```python
sns.barplot(data=tips, x='sex', y='tip', estimator='median')
plt.show()
```

```python
# 흡연여부별 tip의 평균을 시각화
sns.barplot(data=tips, x='smoker', y='tip')
plt.show()
```

```python
# 성별 흡연여부별 tip의 평균을 시각화
sns.barplot(data=tips, x='sex', y='tip', hue='smoker')
plt.show()
```

# 산점도 : `scatterplot()`

* tip ~ total_bill 산점

```python
sns.scatterplot(data=tips, x='tip', y='total_bill')
plt.show()
```

* tip ~ total_bill
* 성별에 따라서 점의 색깔이 다르게 표현

```python
sns.scatterplot(data=tips, x='tip', y='total_bill', hue='sex')
plt.show()
```

* tip ~ total_bill
* 성별에 따라서 점의 색깔을 다르게.
* 흡연 여부에 따라서 점의 모양을 다르게.

```python
sns.scatterplot(data=tips, x='tip', y='total_bill', hue='sex', style='smoker')
plt.show()
```

* tip ~ total_bill
* 성별에 따라서 점의 색깔을 다르게.
* 흡연 여부에 따라서 점의 모양을 다르게.
* 점의 크기를 인원수에 따라서 다르게 표현

```python
sns.scatterplot(data=tips, x='tip', y='total_bill', hue='sex', style='smoker', size='size')
plt.show()
```

# 카테고리 그래프: `catplot()`

*카테고리 별로 subplot들을 생성하고 통계값(평균, 중앙값, ...)을 계산해서, 각각의 subplot에 같은 종류의 그래프들을 시각화 해준다.

```python
# 요일별 팁의 평균을 막대 그래프로 그리기. 흡연 여부도 표현.
sns.barplot(data=tips, x='day', y='tip', hue='smoker')
plt.show()
```

```python
sns.catplot(data=tips, x='day', y='tip', col='smoker', kind='bar')
plt.show()
```

* 요일별 팁의 평균 막대 그래프
* 시간(Lunch, Dinner) 별로 subplot을 생성.

```python
sns.catplot(data=tips, x='day', y='tip', col='time', kind='bar')
plt.show()
```

* 요일별 팁의 평균 막대 그래프
* 흡연여부별로 subplot을 생성
* 시간별 subplot 생성

```python
sns.catplot(data=tips, x='day', y='tip', row='time', col='smoker', kind='bar')
plt.show()
```

* 요일별 성별 팁의 막대 그래프
* 시간별 subplot

```python
sns.catplot(data=tips, x='day', y='tip', hue='sex',col='time', kind='bar')
plt.show()
```

* 성별 팁의 box plot
* 요일별로 sublplot

```python
sns.catplot(data=tips, x='sex', y='tip', col='day', kind='box')
plt.show()
```
