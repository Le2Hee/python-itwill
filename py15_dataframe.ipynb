# 성별에 따른 tips 차이?

* 남성 손님들의 tip의 평균, 여성 손님들의 tip의 평균
* 성별 tip의 최댓값
* 성별 tup의 최솟값

```python
male_tips_mean = tips[tips.sex == 'Male'].tip.mean()
print(f'남성 손님들의 tip의 평균 = {male_tips_mean}')
female_tips_mean = tips[tips.sex == 'Female'].tip.mean()
print(f'여성 손님들의 tip의 평균 = {female_tips_mean}')
male_tips_max = tips[tips.sex == 'Male'].tip.max()
print(f'남성 손님들의 tip의 최댓값 = {male_tips_max}')
female_tips_max = tips[tips.sex == 'Female'].tip.max()
print(f'여성 손님들의 tip의 최댓값 = {female_tips_max}')
male_tips_min = tips[tips.sex == 'Male'].tip.min()
print(f'남성 손님들의 tip의 최솟값 = {male_tips_min}')
female_tips_min = tips[tips.sex == 'Female'].tip.min()
print(f'여성 손님들의 tip의 최솟값 = {female_tips_min}')
```

# 흡연 여부에 따른 tip의 차이?

* 흡연 여부(Yes/No) 별 tip의 평균

```python
smoker_yes = tips[tips.smoker == 'Yes'].tip.mean()
smoker_no = tips[tips.smoker == 'No'].tip.mean()
print(f'흡연자 tip의 평균 = {smoker_yes}')
print(f'비흡연자 tip의 평균 = {smoker_no}')
```

# 요일에 따른 tip의 차이?

* 요일 별 tip의 평균

```python
days = list(set(list(map(str, tips.day))))
days_tips_mean = [tips[tips.day == f'{y}'].tip.mean() for y in days]
tips_day_dict = {
    'day': days,
    'tips_mean': days_tips_mean
}

tdd = pd.DataFrame(tips_day_dict)
tdd
```

# 시간에 따른 tip의 차이?

* 시간 별 tip의 평균

```python
times = list(set(list(map(str, tips.time))))
times_tips_mean = [tips[tips.time == f'{x}'].tip.mean() for x in times]
tips_time_dict = {
    'time': times,
    'tips_mean': times_tips_mean
}

ttd = pd.DataFrame(tips_time_dict)
ttd
```

## 카테고리 별로 2개 이상 변수(컬럼)의 통계값 찾기

```python
# 성별 영수증, 팁의 평균
tips.groupby('sex')[['total_bill', 'tip']].mean()
```

```python
# 성별, 흡연여부별 영수증과 팁의 평균
tips.groupby(['sex', 'smoker'])[['total_bill', 'tip']].mean()
```

## 그룹별 2개 이상의 통계 함수 적용

```python
# 성별 팁의 평균, 최솟값, 중앙값, 최댓값
tips.groupby('sex').tip.agg([np.mean, np.min, np.median, np.max])  # aggregate
```

```python
# 성별, 흡연여부별 영수증과 팁의 평균과 표준편차
tips.groupby(['sex', 'smoker'])[['total_bill', 'tip']].agg([np.mean, np.std])
```

# `pivot_table` 함수(메서드)

* Parameters:
    * values: 통계 함수를 적용할 변수(컬럼)
    * index: 결과 데이터프레임에서 행 인덱스로 사용할 변수(컬럼)
    * columns: 결과 데이터프레임에서 열 인덱스로 사용할 변수(컬럼)
    * aggfunc: aggregation function. 기본값은 평균(mean).

```python
# 성별 팁의 평균
tips.pivot_table(values='tip', index='sex')  # aggfunc='mean' (기본값)
```

```python
# 성별 흡연여부별 팁의 평균
tips.pivot_table(values='tip', index=['sex', 'smoker'])
```

```python
tips.pivot_table(values='tip', index='sex', columns='smoker')
```

```python
# 성별, 흡연여부별, 요일별, 시간별 팁의 평균
# 성별, 흡연여부 - 인덱스. 요일, 시간 - 컬럼
tips.pivot_table(values='tip', index=['sex', 'smoker'], columns=['day', 'time'])
```

```python
# 요일별, 시간별 영수증과 팁의 평균, 표준편차
tips.pivot_table(values=['total_bill', 'tip'],
                 index=['day', 'time'],
                 aggfunc=[np.mean, np.std])
```

```python
tips.pivot_table(values=['total_bill', 'tip'],
                 index='day',
                 columns='time',
                 aggfunc=[np.mean, np.std])
```
