# `matplotlib.pyplot` 모듈을 사용한 시각화

1. `figure` 객체를 생성.
1. `ifgure` 객체에 subplot(axis)을 생성해서 추가.
1. subplot에 그래프 그리기.
1. 그래프의 여러가지 옵션들을 설정.
1. 그래프 보여주기.

```python
# 1. figure 객체 생성
figure = plt.figure()

# 2. 생성된 figure 객체에 subplot(axis)을 추가.
# axis = figure.add_subplot(1, 1, 1) # (행의 개수, 컬럼의 개수, 인덱스) subplot은 인덱스가 0부터가 아니라 1부터 시작한다.
axis = figure.add_subplot(111) # 행의개수, 컬럼의 개수, 인덱스 와 같다.
# figure.add_subplot(212) # 인덱스: 2번째 그래프.

# 3. subplot에 그래프 종류를 선택해서 그린다.
# aixs라는 변수가 들고있다.
axis.hist(x=tips['total_bill'], color='y', bins=30)

# 4. 그래프의 옵션 설정
axis.set_xlabel('Total Bill') # y축 라벨
axis.set_ylabel('Count') # x축 라벨
axis.set_title('Histogram') # title
axis.grid() # 격자

# 그래프 보여주기
plt.show()
```

figure에 2개의 subplot을 상하로 배치하고 그래프 그리기.

```python
# 1. figure 생성
fig = plt.figure(figsize=(6, 8))

# 2. subplot을 figure에 추가
ax1 = fig.add_subplot(2, 1, 1)
ax2 = fig.add_subplot(2, 1, 2)

# 3. 그래프
ax1.hist(x=tips.total_bill, color='y')
ax2.scatter(x=tips.total_bill, y=tips.tip, color='r')

# 4. 옵션
ax1.grid()
ax2.set_ylabel('tip')
ax2.set_xlabel('total_bill')
ax2.grid()

# 5. 보여주기
plt.show()
```

figure에 2개의 subplot 좌우로 배치하고 hist와 scatter 그래프 그리기

```python
fig = plt.figure(figsize=(8, 6))
ax1 = fig.add_subplot(1, 2, 1)
ax2 = fig.add_subplot(1, 2, 2)
ax1.hist(x=tips.total_bill, color='y')
ax2.scatter(x=tips.total_bill, y=tips.tip, color='r')
ax1.grid()
ax2.grid()
plt.show()
```

# `plt.subplots()` 함수

figure 생성 + subplot 추가 (항상 1번 2번으로 일어나야 된다.)

## figure에 1개의 subplot이 있는 경우

```python
# figure생성, figure에 subplot을 추가.
fig, ax = plt.subplots()

# subplot에 그래프 그리기.
ax.hist(x=tips.total_bill)

# 그래프 보여주기
plt.show()
```

## subplot 2개를 상하로 배치

```python
fig, ax = plt.subplots(nrows=2, figsize=(6, 8))
# print(ax) #  subplot이 2개 이상인 경우 리턴 값 ax는 subplot들의 배열(ndarray)

ax[0].hist(x=tips.total_bill)
ax[1].scatter(x=tips.total_bill, y=tips.tip)

plt.show()
```

## 2개의 subplot을 좌우로 배

```python
fig, ax = plt.subplots(ncols=2, figsize=(12, 4))
ax[0].hist(x=tips.total_bill)
ax[1].scatter(x=tips.total_bill, y=tips.tip)
plt.show()
```

# `pyplot` 모듈의 top-level 함수

figure 생성 + subplot 1개 생성 + 그래프 종

## Histogram

* 연속형 변수를 구간으로 나눠서, 그 구간 안에 포함된 자료의 개수를 막대의 높이로 표현한 그래프.
* 연속형 변수의 분포를 알 수 있다.

hist 함수 파라미터:

* x: x축에 사용할 변수 이름. 히스토그램을 그리기 위한 자료.
* bins: 막대 개수, 구간
* color: 막대 안쪽을 채우는 색깔
* edgecolor: 막대 테두리 색깔
* density: False -> 개수, True -> 밀도(비율)

```python
# 1. figure 생성, 2. subplot 1개 생성, 3. histogram 그리기
plt.hist(x=tips.total_bill, bins=40, color='g', edgecolor='y', density=True)

# 4. 옵션 설정 -> 함수 이름이 달라진다.
plt.xlabel("Total bill")
plt.ylabel("Count")
plt.title("Histogram")

# 5. 보여주기
plt.show()
```

## Bar plot(막대 그래프)

카테고리 타입 변수의 빈도수를 막대로 표현한 그래프.



```python
gender_count = tips.sex.value_counts()
gender_count
```

```python
type(gender_count)
```

```python
gender_count.index
```

```python
gender_count.values
```

```python
#세로 막대 그래프(vertical bar plot)
# x: 인덱스, height: 막대의 높이, 빈도수.
plt.bar(x=gender_count.index, height=gender_count.values, width=0.2)
plt.show()
```

```python
# 가로 막대 그래프(horizontal bar graph)
plt.barh(y=gender_count.index, width=gender_count.values, height=0.4)
plt.show()
```

### Ex4. 성별 tip의 box plot을 하나의 subplot에 그리기

```python
sex_index = tips.sex.value_counts().index
male_tips = tips[tips['sex'] == 'Male'].tip
female_tips = tips[tips['sex'] == 'Female'].tip
plt.boxplot(x=[male_tips, female_tips])
plt.xticks([1, 2],[sex_index[0], sex_index[1]])
plt.show()
```

### Ex5. 시간별 tip의 box plot을 하나의 subplot에 그리기

### Ex6. 요일별 tip의 box plot을 하나의 subplot에 그리기

```python
day_index = tips.day.value_counts().index
day_index
sat_count = tips[tips.day == 'Sat'].tip
sun_count = tips[tips.day == 'Sun'].tip
thur_count = tips[tips.day == 'Thur'].tip
fri_count = tips[tips.day == 'Fri'].tip
plt.boxplot(x=[sat_count, sun_count, thur_count, fri_count])
plt.xticks([1, 2, 3, 4], [day_index[0], day_index[1],day_index[2], day_index[3]])
plt.show()
```
