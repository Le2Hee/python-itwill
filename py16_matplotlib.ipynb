# `matplotlib.pyplot` 모듈을 사용한 시각화

1. `figure` 객체를 생성.
1. `ifgure` 객체에 subplot(axis)을 생성해서 추가.
1. subplot에 그래프 그리기.
1. 그래프의 여러가지 옵션들을 설정.
1. 그래프 보여주기.

```python
# 1. figure 객체 생성
figure = plt.figure()

# 2. 생성된 figure 객체에 subplot(axis)을 추가.
# axis = figure.add_subplot(1, 1, 1) # (행의 개수, 컬럼의 개수, 인덱스) subplot은 인덱스가 0부터가 아니라 1부터 시작한다.
axis = figure.add_subplot(111) # 행의개수, 컬럼의 개수, 인덱스 와 같다.
# figure.add_subplot(212) # 인덱스: 2번째 그래프.

# 3. subplot에 그래프 종류를 선택해서 그린다.
# aixs라는 변수가 들고있다.
axis.hist(x=tips['total_bill'], color='y', bins=30)

# 4. 그래프의 옵션 설정
axis.set_xlabel('Total Bill') # y축 라벨
axis.set_ylabel('Count') # x축 라벨
axis.set_title('Histogram') # title
axis.grid() # 격자

# 그래프 보여주기
plt.show()
```

figure에 2개의 subplot을 상하로 배치하고 그래프 그리기.

```python
# 1. figure 생성
fig = plt.figure(figsize=(6, 8))

# 2. subplot을 figure에 추가
ax1 = fig.add_subplot(2, 1, 1)
ax2 = fig.add_subplot(2, 1, 2)

# 3. 그래프
ax1.hist(x=tips.total_bill, color='y')
ax2.scatter(x=tips.total_bill, y=tips.tip, color='r')

# 4. 옵션
ax1.grid()
ax2.set_ylabel('tip')
ax2.set_xlabel('total_bill')
ax2.grid()

# 5. 보여주기
plt.show()
```

figure에 2개의 subplot 좌우로 배치하고 hist와 scatter 그래프 그리기

```python
fig = plt.figure(figsize=(8, 6))
ax1 = fig.add_subplot(1, 2, 1)
ax2 = fig.add_subplot(1, 2, 2)
ax1.hist(x=tips.total_bill, color='y')
ax2.scatter(x=tips.total_bill, y=tips.tip, color='r')
ax1.grid()
ax2.grid()
plt.show()
```

## Histogram

* 연속형 변수를 구간으로 나눠서, 그 구간 안에 포함된 자료의 개수를 막대의 높이로 표현한 그래프.
* 연속형 변수의 분포를 알 수 있다.

hist 함수 파라미터:

* x: x축에 사용할 변수 이름. 히스토그램을 그리기 위한 자료.
* bins: 막대 개수, 구간
* color: 막대 안쪽을 채우는 색깔
* edgecolor: 막대 테두리 색깔
* density: False -> 개수, True -> 밀도(비율)

```python
# 1. figure 생성, 2. subplot 1개 생성, 3. histogram 그리기
plt.hist(x=tips.total_bill, bins=40, color='g', edgecolor='y', density=True)

# 4. 옵션 설정 -> 함수 이름이 달라진다.
plt.xlabel("Total bill")
plt.ylabel("Count")
plt.title("Histogram")

# 5. 보여주기
plt.show()
```

# 산점도(Scatter plot)

두 변수(컬럼) 사이의 상관 관계를 파악하기 위해서 사용.

* 점들을 그린다.

```python
# tip ~ total_bill 상관 관계
plt.scatter(x=tips.total_bill, y=tips.tip,
            alpha=0.3,  # alpha: 0(투명) ~ 1.0(불투명)
            s=20,  # s: size, 마커(점)의 크기.
            c='g',  # c: color, 마커(점)의 색상.
            marker='*'  # marker: 마커의 종류(모양), 'o': 동그라미 모양, '^': 삼각형, 'v': 역삼각형, '*': 별모양, 's': 정사각형, ...
            )

plt.grid()
plt.xlabel('Total bill')
plt.ylabel('Tip')

plt.show()
```

```python
# tip ~ size 상관 관계
plt.scatter(x=tips['size'], y=tips.tip,
            alpha=0.3,
            s=40,
            c='r',
            marker='^'
            )
plt.grid()
plt.xlabel('Size')
plt.ylabel('Tip')

plt.show()
```

## 선 그래프(Line plot)

주로 시계열 데이터(시간에 따라서 변하는 데이터: 주가, 환율, 날씨, ...)를 시각화.

```python
np.random.seed(1) # 난수 생성하는 순서.
df = pd.DataFrame(data={
    'time': np.arange(1, 11), # [1, 2, 3, 4, ..., 9, 10]
    'price': np.random.rand(10) # [0, 1.0) 범위의 난수 10개를 갖는 1차원 배열.
})
df
```

```python
# price ~ time
plt.plot(df.time, df.price,
        'k+--' # 모양 '-'를 사용하면 마커도 보이고 선도 보인다.
        )
# format 문자열(3번째 arg): 색깔, 마커모양, 선의 종류 로 이루어져 있다.
# 색깔: b(lue), r(de), g(reen), (blac)k, ...
# marker: 마커의 종류(모양), 'o': 동그라미 모양, '^': 삼각형, 'v': 역삼각형, '*': 별모양, 's': 정사각형, ...
# -(solid line), :(dotted line), --(dashed line), ...

plt.grid()
plt.xlabel('Time')
plt.ylabel('Price')

plt.show()
```

```python
# variable-length keyword argument
plt.plot(df.time, df.price,
         c='crimson', # 선, 마커 색상.
         marker='d', # 마커 모양
         linestyle='dotted', # 선 모양
         markersize=20,

        )

plt.show()
```

1개의 subplot에 선 그래프 3개를 중첩시켜서 표현.

```python
np.random.seed(1)
df = pd.DataFrame(data={
    'time': np.arange(1, 11),
    'price1': np.random.rand(10),
    'price2': np.random.rand(10),
    'price3': np.random.rand(10)
})
df
```

```python
# price1 ~ time
plt.plot(df.time, df.price1, 'rP--', label='price1')
plt.plot(df.time, df.price2, label='price2', c='pink', marker='d', linestyle='dotted')
plt.plot(df.time, df.price3, 'y3:', label='price3')

plt.legend() # 범례, 카테고리의 설명, 화면에 표시
plt.grid()
plt.xlabel('Time')
plt.ylabel('Price')

plt.show()
```
